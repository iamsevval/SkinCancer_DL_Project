```python
dst_dir = "/Users/eylul/Desktop/all_images"

label_map = {
    "akiec": 0,
    "bcc": 1,
    "bkl": 2,
    "df": 3,
    "mel": 4,
    "nv": 5,
    "vasc": 6
}

data_list = []

for item in metadata:
    img_id = item["image_id"]
    label = label_map[item["dx"]]
    img_path = os.path.join(dst_dir, img_id + ".jpg")

    if os.path.exists(img_path):
        data_list.append((img_path, label))

print(f"Toplam kullanƒ±labilir veri: {len(data_list)}")
```

```python
import random


class_groups = {i: [] for i in range(7)}
for path, label in data_list:
    class_groups[label].append((path, label))

train_list, val_list, test_list = [], [], []

for label, items in class_groups.items():
    random.shuffle(items)
    n = len(items)
    test_n = int(n * 0.15)
    val_n = int(n * 0.15)
    test_list.extend(items[:test_n])
    val_list.extend(items[test_n:test_n+val_n])
    train_list.extend(items[test_n+val_n:])


random.shuffle(train_list)
random.shuffle(val_list)
random.shuffle(test_list)

print(f"Train: {len(train_list)}, Val: {len(val_list)}, Test: {len(test_list)}")
```

```python
import csv
import os
from sklearn.model_selection import train_test_split
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image

# üîπ 1Ô∏è‚É£ DOƒûRU dosya ve resim klas√∂r√º yollarƒ±
csv_file = "/Users/eylul/Desktop/all_images/HAM10000_metadata.csv"
image_dir = "/Users/eylul/Desktop/all_images"

# üîπ 2Ô∏è‚É£ CSV‚Äôden veriyi oku ve tam resim yolunu olu≈ütur
data_list = []
with open(csv_file, newline='', encoding="utf-8") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        image_filename = row['image_id'] + ".jpg"
        image_path = os.path.join(image_dir, image_filename)
        label = row['dx']
        data_list.append({
            'image_path': image_path,
            'label': label
        })

print("Toplam veri:", len(data_list))

# üîπ 3Ô∏è‚É£ Train / Validation / Test b√∂lme
train_list, temp_list = train_test_split(
    data_list,
    test_size=0.3,
    random_state=42,
    stratify=[d['label'] for d in data_list]
)

val_list, test_list = train_test_split(
    temp_list,
    test_size=0.5,
    random_state=42,
    stratify=[d['label'] for d in temp_list]
)

print("Train:", len(train_list), "Val:", len(val_list), "Test:", len(test_list))

# üîπ 4Ô∏è‚É£ Dataset sƒ±nƒ±fƒ±
class SkinDataset(Dataset):
    def __init__(self, data_list, transform=None):
        self.data_list = data_list
        self.transform = transform

    def __len__(self):
        return len(self.data_list)

    def __getitem__(self, idx):
        item = self.data_list[idx]
        image = Image.open(item['image_path']).convert("RGB")
        label = item['label']
        if self.transform:
            image = self.transform(image)
        return image, label

# üîπ 5Ô∏è‚É£ Transformlar
train_transform = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
])

val_test_transform = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(),
])

# üîπ 6Ô∏è‚É£ Dataset & DataLoader
train_dataset = SkinDataset(train_list, transform=train_transform)
val_dataset = SkinDataset(val_list, transform=val_test_transform)
test_dataset = SkinDataset(test_list, transform=val_test_transform)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

print("‚úÖ DataLoader hazƒ±r!")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Kullanƒ±lan cihaz:", device)
```

```python
from torchvision import models
import torch.nn as nn

num_classes = 7

# 1Ô∏è‚É£ Modeli olu≈ütur
model = models.efficientnet_b0(weights=None)  # offline

# 2Ô∏è‚É£ Feature extractor'ƒ± dondur (TRANSFER LEARNING)
for param in model.features.parameters():
    param.requires_grad = False

# 3Ô∏è‚É£ Classifier'ƒ± deƒüi≈ütir
in_features = model.classifier[1].in_features
model.classifier[1] = nn.Linear(in_features, num_classes)

# 4Ô∏è‚É£ Cihaza g√∂nder
model = model.to(device)
```

```python
# üîí Feature extractor'ƒ± dondur
for param in model.features.parameters():
    param.requires_grad = False

# üîì Sadece classifier a√ßƒ±k
for param in model.classifier.parameters():
    param.requires_grad = True

optimizer = optim.Adam(
    model.classifier.parameters(),
    lr=1e-4
)
```

```python
# Loss fonksiyonu
criterion = nn.CrossEntropyLoss()

# Optimizer (SADECE classifier eƒüitilecek ‚Üí transfer learning)
optimizer = optim.Adam(
    model.classifier.parameters(),
    lr=0.0001
)

print("‚úÖ Criterion ve optimizer hazƒ±r")
```

```python
num_epochs = 10

train_losses, val_losses = [], []
train_accs, val_accs = [], []

for epoch in range(num_epochs):

    # ================= TRAIN =================
    model.train()
    if epoch == 5:
        print("üîì Fine-tuning ba≈ülatƒ±ldƒ±")

        # Son feature bloƒüunu a√ß
        for param in model.features[-1].parameters():
            param.requires_grad = True

        # Optimizer'ƒ± g√ºncelle (d√º≈ü√ºk LR)
        optimizer = optim.Adam(
            filter(lambda p: p.requires_grad, model.parameters()),
            lr=1e-5
        )
    running_loss = 0.0
    correct = 0
    total = 0

    for batch_idx, (images, labels) in enumerate(train_loader):
        images = images.to(device)

        # üî• LABEL D√ñN√ú≈û√úM√ú (string ‚Üí int)
        labels = torch.tensor(
            [label_map[l] for l in labels],
            dtype=torch.long
        ).to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, preds = torch.max(outputs, 1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

        # üîπ DONDU MU G√ñR
        if batch_idx % 50 == 0:
            print(f"Epoch {epoch+1}/{num_epochs} | Train batch {batch_idx}/{len(train_loader)}")

    train_losses.append(running_loss / len(train_loader))
    train_accs.append(correct / total)

    # ================= VALIDATION =================
    model.eval()
    running_loss = 0.0
    correct = 0
    total = 0

    with torch.no_grad():
        for images, labels in val_loader:
            images = images.to(device)
            labels = torch.tensor(
                [label_map[l] for l in labels],
                dtype=torch.long
            ).to(device)

            outputs = model(images)
            loss = criterion(outputs, labels)

            running_loss += loss.item()
            _, preds = torch.max(outputs, 1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)

    val_losses.append(running_loss / len(val_loader))
    val_accs.append(correct / total)

    print(
        f"‚úÖ Epoch [{epoch+1}/{num_epochs}] | "
        f"Train Acc: {train_accs[-1]:.3f} | "
        f"Val Acc: {val_accs[-1]:.3f}"
    )
```

```python
# ====== Kanser / Kanser Deƒüil Haritasƒ± ======

cancer_map = {
    "akiec": "Kanser",
    "bcc": "Kanser",
    "mel": "Kanser",
    "nv": "Kanser Deƒüil",
    "bkl": "Kanser Deƒüil",
    "df": "Kanser Deƒüil",
    "vasc": "Kanser Deƒüil"
}

inv_label_map = {v: k for k, v in label_map.items()}
```

```python
# ================= Eƒûƒ∞Tƒ∞M GRAFƒ∞KLERƒ∞ =================
import matplotlib.pyplot as plt

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.plot(train_accs, label="Train Accuracy")
plt.plot(val_accs, label="Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.title("EfficientNet - Accuracy")

plt.subplot(1,2,2)
plt.plot(train_losses, label="Train Loss")
plt.plot(val_losses, label="Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.title("EfficientNet - Loss")

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import roc_auc_score
import torch.nn.functional as F

model.eval()

all_preds = []
all_labels = []
all_probs = [] # Olasƒ±lƒ±klarƒ± saklamak i√ßin yeni liste

with torch.no_grad():
    for images, labels in test_loader:
        images = images.to(device)

        labels = torch.tensor(
            [label_map[l] for l in labels],
            dtype=torch.long
        ).to(device)

        outputs = model(images)
        
        # Olasƒ±lƒ±klarƒ± hesapla (Softmax)
        probs = F.softmax(outputs, dim=1)
        
        _, preds = torch.max(outputs, 1)

        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())
        all_probs.extend(probs.cpu().numpy()) # Olasƒ±lƒ±klarƒ± ekle

# ROC-AUC Hesaplama (One-vs-Rest)
roc_auc = roc_auc_score(all_labels, all_probs, multi_class='ovr')

print(f"‚úÖ Test tamamlandƒ±")
print(f"üìä EfficientNet ROC-AUC Skoru: {roc_auc:.4f}")
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(all_labels, all_preds)

plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Tahmin")
plt.ylabel("Ger√ßek")
plt.title("Confusion Matrix - EfficientNet")
plt.show()

print(classification_report(all_labels, all_preds, zero_division=0))
```
